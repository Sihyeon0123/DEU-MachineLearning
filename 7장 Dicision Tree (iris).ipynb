```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib import pyplot as plt
```

```python
# 데이터 로딩
iris = load_iris()
iris
```

```python
# 입력 항목명
print(iris.feature_names)
# 목표 클래스의 유형 (이름)
print(iris.target_names)
```

```python
# 입력 데이터 세트
X = iris.data
# 목표 데이터 세트
Y = iris.target
```

```python
# 입력 데이터 세트의 행과 열의 크기 출력
print(X.shape)
# 입력 데이터 세트 출력
print(X)
```

```python
# 목표 데이터 세트의 행과 열의 크기 출력
print(Y.shape)
# 목표 데이터 세트 출력
print(Y)
```

```python
# 데이터 세트의 분리ㅣ 학습용 입력값, 테스트용 입력값, 학습용 목표값, 테스트용 목표값
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=1234)
```

```python
# 학습용 이ㅏㅂ력 데이터 세트의 행과 열의 크기 출력
print(X_train.shape)
# 학습용 입력값 출력
print(X_train)
```

```python
# 학습용 목표 데이터 세트의 행과 열의 크기 출력
print(Y_train.shape)
# 학습용 목표값의 출력
print(Y_train)
```

```python
# 테스트용 입력 데이트 세트의 행과 열의 크기 출력
print(X_test.shape)
# 테스트용 입력값의 출력
print(X_test)
```

```python
# 테스트용 목표 데이터 세트의 행과 열의 크기 출력
print(Y_test.shape)
# 테스트용 목표값의 출력
print(Y_test)
```

```python
# C5.0 모형 설정
clf = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
# 학습
clf = clf.fit(X_train, Y_train)
```

```python
# 그래프 출력 영역 (가로와 세로, 단위는 인치)
plt.figure(figsize=(15,12))
# 학습된 모형 출력
tree.plot_tree(clf)
plt.show()
```

```python
# 분류 예측
Y_pred = clf.predict(X_test)
print(Y_pred)
```

```python
# 평가: 혼동 행렬 출력
print(confusion_matrix(Y_test, Y_pred))
```

```python
# 정확도 출력
print("정확도:", accuracy_score(Y_test, Y_pred))
```
